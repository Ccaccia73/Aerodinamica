```python
import numpy as np
import matplotlib.pyplot as plt
import glob
%matplotlib inline
```

```python
inv_cli = glob.glob('CLi_NACA_*.dat')
```

```python
inv_cli
```

```python
CLi = {}
```

```python
for fname in inv_cli:
    CLi[fname[4:8]+' '+fname[9:13]] = np.genfromtxt(fname,skip_header=12)
```

```python
plt.figure(figsize=(16,10))
for foil in CLi:
    plt.plot(CLi[foil][:,0],CLi[foil][:,1],label=foil)
    plt.xlabel(r'angle [deg]')
    plt.ylabel(r'$C_l$')
    plt.legend()
    plt.grid(True)
    plt.title('Inviscid lift coefficient')
```

```python
foil = '23012' # 1412 2212 2412 23012
```

```python
coord_naca_1412 = np.genfromtxt('NACA_'+foil+'.dat',skip_header=1)
```

```python
plt.figure(figsize=(16,4))
plt.plot(coord_naca_1412[:,0],coord_naca_1412[:,1])
plt.title('Profile NACA '+foil)
plt.axis('equal');
```

```python
CPi_NACA_1412 = glob.glob('CPi_NACA_'+foil+'*.dat')
```

```python
CPi_1412 = {}
for fname in CPi_NACA_1412:
    CPi_1412[int(fname.split('_')[-1].split('.')[0])] = np.genfromtxt(fname,skip_header=1)
```

```python
plt.figure(figsize=(16,10))
for angle in sorted(CPi_1412):
    if angle < 0:
        style = ':'
        sz = 2.5
    elif angle < 2:
        style = '-.'
        sz = 2
    elif angle < 4:
        style = '--'
        sz = 3
    else:
        style = '-'
        sz = 4
    plt.plot(CPi_1412[angle][:,0],CPi_1412[angle][:,1],label=str(angle),ls=style,lw=sz)
    plt.xlabel(r'$\frac{x}{c}$')
    plt.ylabel(r'$Cp$')
    plt.legend()
    plt.gca().invert_yaxis()
    plt.grid(True)
    plt.title('Pressure coefficient for NACA '+foil)
```
