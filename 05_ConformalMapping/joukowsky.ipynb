```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 11
```

```python
def joukowsky_transform(ζ,c):
    return ζ + c**2/ζ
```

```python
N = 250
θ = np.linspace(0,2*np.pi,N)
```

```python
ζ0 = np.cos(θ) +1j*np.sin(θ)
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
plt.plot(ζ0.real,ζ0.imag)
plt.axis('equal')
z0 = joukowsky_transform(ζ0,1)
plt.subplot(1,2,2)
plt.plot(z0.real,z0.imag)
plt.axis('equal');
```

```python
ζ1 = 1.2*(np.cos(θ) +1j*np.sin(θ))
z1 = joukowsky_transform(ζ1,1)
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
plt.plot(ζ1.real,ζ1.imag)
plt.axis('equal')
plt.grid()
plt.subplot(1,2,2)
plt.plot(z1.real,z1.imag)
plt.axis('equal')
plt.grid();
```

```python
ζ2 = 5/6*(np.cos(θ) +1j*np.sin(θ))
z2 = joukowsky_transform(ζ2,1)
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
plt.plot(ζ2.real,ζ2.imag)
plt.axis('equal')
plt.grid()
plt.subplot(1,2,2)
plt.plot(z2.real,z2.imag)
plt.axis('equal')
plt.grid();
```

```python
R = 0.9
c0 = 0.1
L = 1
ζ3 = R*((np.cos(θ) +1j*np.sin(θ)))+c0
z3 = joukowsky_transform(ζ3,1)
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
plt.plot(ζ3.real,ζ3.imag)
plt.plot(c0.real,c0.imag,'ro')
plt.plot([-L,L],[0,0],'gx')
plt.axis('equal')
plt.grid()
plt.subplot(1,2,2)
plt.plot(z3.real,z3.imag)
plt.axis('equal')
plt.grid();
```

```python
c0 = 0.5j
L = 1
R = np.absolute(L+c0)
ζ3 = R*((np.cos(θ) +1j*np.sin(θ)))+c0
z3 = joukowsky_transform(ζ3,1)
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
plt.plot(ζ3.real,ζ3.imag)
plt.plot(c0.real,c0.imag,'ro')
plt.plot([-L,L],[0,0],'gx')
plt.axis('equal')
plt.grid()
plt.subplot(1,2,2)
plt.plot(z3.real,z3.imag)
plt.axis('equal')
plt.grid();
```
