```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 11
```

```python
def joukowsky_transform(ζ,c):
    return ζ + c**2/ζ
```

```python
δ = np.logspace(-7,0,num=250)
R = np.sqrt(1+δ**2)

p_sup = joukowsky_transform((δ+R)*1j,1)
p_inf = joukowsky_transform((δ-R)*1j,1)
```

```python
plt.figure(figsize=(16,8))
#plt.semilogx(δ,p_sup.real,label="Re sup")
plt.semilogx(δ,p_sup.imag,label="Im sup")
#plt.semilogx(δ,p_inf.real,label="Re inf")
plt.semilogx(δ,p_inf.imag,label="Im inf")
```

```python
np.max(p_sup.imag - p_inf.imag)
```
